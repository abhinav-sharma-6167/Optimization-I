### Question C

Lehmer matrix of dimension 20




```python
def lehmer_entry(x=20,y=20):
    X = np.zeros((x,y))
    for i in range(x):
        for j in range(y):
            if (j > i):
                X[i,j] = (i+1)/(j+1)
            else :
                X[i,j] = (j+1)/(i+1)
    return(X)
```

```python
lehmer_entry(x=4,y=4)
```

```python
A = lehmer_entry(x=20,y=20)
A
```
